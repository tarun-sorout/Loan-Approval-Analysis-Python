```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/1loan_approval.csv')
df.head()
```

```python
print(df.info())
display(df.describe())
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Approval Status', data=df, palette='viridis', hue='Approval Status', legend=False)
plt.title('Distribution of Loan Approval Status')
plt.xlabel('Approval Status')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
numerical_features = ['Income', 'Credit Score', 'Loan Amount']

plt.figure(figsize=(18, 5))
for i, feature in enumerate(numerical_features):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(x='Approval Status', y=feature, data=df, palette='muted', hue='Approval Status', legend=False)
    plt.title(f'{feature} vs. Approval Status')
    plt.xlabel('Approval Status')
    plt.ylabel(feature)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Convert 'Approval Status' to numerical for correlation analysis
df_encoded = df.copy()
df_encoded['Approval Status'] = df_encoded['Approval Status'].map({'Rejected': 0, 'Approved': 1})

correlation_matrix = df_encoded[['Income', 'Credit Score', 'Loan Amount', 'Approval Status']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features and Approval Status')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Income', hue='Approval Status', kde=True, palette='viridis', alpha=0.7)
plt.title('Distribution of Income by Approval Status')
plt.xlabel('Income')
plt.ylabel('Density / Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Credit Score', hue='Approval Status', kde=True, palette='magma', alpha=0.7)
plt.title('Distribution of Credit Score by Approval Status')
plt.xlabel('Credit Score')
plt.ylabel('Density / Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Loan Amount', hue='Approval Status', kde=True, palette='cividis', alpha=0.7)
plt.title('Distribution of Loan Amount by Approval Status')
plt.xlabel('Loan Amount')
plt.ylabel('Density / Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
